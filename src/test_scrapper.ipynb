```python
import json
import pandas as pd
```

```python
from jobstreet_scrapper import JobStreetScrapper
from linkedin_scrapper import LinkedInScrapper
```

```python
# parse filters
def read_config(file_path):
    with open(file_path, 'r') as file:
        config = json.load(file)
    return config
```

```python
job = 'Software Engineer'
config = read_config('config.json')
```

```python
jobstreet_jobs = JobStreetScrapper.get_jobs(job, config)
jobstreet_jobs
```

```python
linkedin_jobs = LinkedInScrapper.get_jobs(job, config)
linkedin_jobs
```

```python
jobs_df = pd.concat([jobstreet_jobs, linkedin_jobs]).reset_index(drop=True)
jobs_df
```

```python
jobs_df.to_csv('jobs.csv')  
```
